# Welcome

This short introduction is designed to get you up and running notebooks. What's a notebook? You're looking at one! It's a mixture of text (like this text you're reading) and code like this:

```python
1+1
```

Notebooks can also do visual things like plot a graph:

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.show()
```

## Start a New Notebook

1. Open a new browser window on the [notebooks home page here.](https://notebooks.geobigdata.io/)
2. You may need to log in.
3. Once you're on the notebooks home page, click "+ Create New" at the top. Click to create

You now have a new, blank, notebook.

## Writing Code

At the top of your notebook, below the menus and toolbars, you'll see a text box with "In: [ ]" on the left. Click on that text box, and enter the text `2+2`.

Now, to execute that code hit `CTRL-Enter`

The result should look like this:

```python
2+2
```

## Writing Text

The wonderful thing about notebooks is that you can mix code and text at the same time. To add text, first add a new cell with `Insert -> Insert Cell Below`. Then convert the cell to text with `Cell -> Cell Type -> Markdown`. You can edit the text by double-clicking the cell and then editing using markdown (here's a [cheatsheet](https://github.com/adam-p/markdown-here/wiki/Markdown-Cheatsheet)). To render the text, execute the cell as you normally would with `CTRL-Enter`.

## Imagery

DigitalGlobe's archive of satellite imagery spans the whole world over many years. We can pull imagery in to a notebook with code, and there's a helper on the left to make this easier.

Click the 'imagery' tab on the left of a notebook and then pan and zoom the map to find an area you are interested in. Click the rectangle tool below the pentagon icon on the map, and drag an area of interest. Note - these tools only show when you are zoomed in to small area like atown or city or smaller.

Drag a green box using that Area of Interest (AoI). Then below, hit 'Search' to find imagery.
